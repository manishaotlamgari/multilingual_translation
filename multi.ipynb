```python
import os
import json
import speech_recognition as sr
import noisereduce as nr
import numpy as np
import wave
from googletrans import Translator  # Use Google Translate API
from gtts import gTTS
import IPython.display as ipd
import ipywidgets as widgets
from IPython.display import display
from google.colab import files
from pydub import AudioSegment

# Load language codes from JSON file
with open('language_codes.json', 'r') as f:
    language_dict = json.load(f)  # Dictionary: {"English": "en", "French": "fr", ...}

# UI Elements
upload_button = widgets.FileUpload(accept=".wav,.mp3", multiple=False)
language_dropdown = widgets.SelectMultiple(
    options=language_dict.keys(), description="Languages:", layout={'width': 'max-content'}
)
status_label = widgets.Label(value="Status: Ready")
spoken_text_area = widgets.Textarea(description="Spoken Text", layout={'width': '100%', 'height': '100px'})
translated_text_areas = {}  # Dictionary to hold text areas for each language
download_buttons = {}  # Dictionary to hold download buttons

# Function to reduce noise
def reduce_noise(file_name):
    with wave.open(file_name, "rb") as wf:
        params = wf.getparams()
        audio_data = wf.readframes(wf.getnframes())
        audio_array = np.frombuffer(audio_data, dtype=np.int16)
        reduced_noise = nr.reduce_noise(y=audio_array, sr=params.framerate)

    # Save cleaned audio
    cleaned_file = "cleaned_audio.wav"
    with wave.open(cleaned_file, "wb") as wf:
        wf.setparams(params)
        wf.writeframes(reduced_noise.astype(np.int16).tobytes())

    return cleaned_file

# Function to process uploaded audio
def process_audio(change):
    status_label.value = "Status: Processing file..."

    # Get uploaded file
    uploaded_file = next(iter(upload_button.value))
    file_name = uploaded_file
    with open(file_name, 'wb') as f:
        f.write(upload_button.value[file_name]['content'])

    # Convert MP3 to WAV if necessary
    if file_name.endswith('.mp3'):
        audio = AudioSegment.from_mp3(file_name)
        file_name = "converted_audio.wav"
        audio.export(file_name, format="wav")

    # Reduce noise
    file_name = reduce_noise(file_name)

    # Speech recognition
    recognizer = sr.Recognizer()
    with sr.AudioFile(file_name) as source:
        audio_data = recognizer.record(source)

    try:
        spoken_text = recognizer.recognize_google(audio_data)
        spoken_text_area.value = spoken_text
        status_label.value = "Status: Recognized Speech"

        # Translate text into selected languages
        translator = Translator()
        selected_languages = language_dropdown.value

        # Remove old text areas and buttons
        for lang in translated_text_areas:
            translated_text_areas[lang].close()
            download_buttons[lang].close()

        translated_text_areas.clear()
        download_buttons.clear()

        for lang in selected_languages:
            lang_code = language_dict[lang]
            translated_text = translator.translate(spoken_text, dest=lang_code).text

            # Create a text area for translated text
            translated_text_area = widgets.Textarea(
                value=translated_text, description=f"{lang}:", layout={'width': '100%', 'height': '100px'}
            )
            translated_text_areas[lang] = translated_text_area
            display(translated_text_area)

            # Convert translated text to speech and save
            tts = gTTS(translated_text, lang=lang_code)
            audio_filename = f"translated_audio_{lang_code}.mp3"
            tts.save(audio_filename)

            # Create a download button for the audio file
            def download_audio(file_name):
                def func(b):
                    files.download(file_name)
                return func

            download_button = widgets.Button(description=f"Download {lang} Audio", button_style="primary")
            download_button.on_click(download_audio(audio_filename))
            download_buttons[lang] = download_button
            display(download_button)

        status_label.value = "Status: Translation & Audio Ready"

    except Exception as e:
        status_label.value = f"Error: {e}"

# Bind upload button
upload_button.observe(process_audio, names='value')

# Display UI
display(upload_button, language_dropdown, status_label, spoken_text_area)
```
